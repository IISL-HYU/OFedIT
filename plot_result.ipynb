```python
import pickle
import matplotlib.pyplot as plt

with open("./Accuracy_lists/OFedAvg.pkl","rb") as f:
    accuracy_list = pickle.load(f)

with open("./Accuracy_lists/OFedPA.pkl","rb") as f:
    p_accuracy_list = pickle.load(f)

with open("./Accuracy_lists/OFedPA2.pkl","rb") as f:
    p2_accuracy_list = pickle.load(f)
    
t1 = list(range(len(accuracy_list)))
t2 = list(range(len(p_accuracy_list)))
t3 = list(range(len(p2_accuracy_list)))

y1 = accuracy_list
y2 = p_accuracy_list
y3 = p2_accuracy_list

# plotting the points
plt.plot(t1, y1, 'black', label='OFedAvg')
plt.plot(t2, y2, 'r-', label=r'OFedPA ($\mathdefault{p=1}$, $\mathdefault{L=3}$)')
plt.plot(t3, y3, 'b-', label=r'OFedPA ($\mathdefault{p=1}$, $\mathdefault{L=10}$)')

plt.ylim([0, 1]) 
plt.legend(ncol = 1)

# naming the x axis
plt.xlabel('t')
# naming the y axis
plt.ylabel('Classification Accuracy')
 
# giving a title to my graph
 
# function to show the plot
plt.grid()
plt.savefig('./Figures/full_part_L_3_L_10.png', dpi=200, facecolor="white")
plt.show()

```
